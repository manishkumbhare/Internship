# 1) Write a python program to display all the header tags from wikipedia.org and make data frame.

```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
```

```python
page= requests.get('https://en.wikipedia.org/wiki/Main_Page')
page
```

```python
soup= BeautifulSoup(page.content)
```

```python
headers = []
for i in soup.find_all('span',class_="mw-headline"):
    headers.append(i.text)
headers
```

```python
data = pd.DataFrame(headers)
data
```

# 3) Write a python program to scrape cricket rankings from icc-cricket.com. You have to scrape and make data frame

# a) Top 10 ODI teams in men’s cricket along with the records for matches, points and rating

```python
page= requests.get('https://www.icc-cricket.com/rankings/mens/team-rankings/odi')
page
```

```python
soup = BeautifulSoup(page.content)
```

```python
team = []
matches = []
points = []
ratings = []

for j in soup.find_all('tr',class_="rankings-block__banner"):
        matches.append(j.text.split('\n')[7])
        points.append(j.text.split('\n')[8])
        ratings.append(j.text.split('\n')[10].replace(' ',''))
        
for i in soup.find_all('tr',class_="table-body"):
    matches.append(i.text.split('\n')[7])
    points.append(i.text.split('\n')[8])
    ratings.append(i.text.split('\n')[9])
    
for i in soup.find_all('span',class_="u-hide-phablet"):
    team.append(i.text)

data = pd.DataFrame({"ODI Teams" : team, "Matches" : matches , "Points": points, "Ratings": ratings})

data[:10]
```

# 5) Write a python program to scrape mentioned news details from https://www.cnbc.com/world/?region=world and make data frame

# i) Headline

```python
page = requests.get('https://www.cnbc.com/world/?region=world')
soup= BeautifulSoup(page.content)
```

```python
head_lines= []
for i in soup.find_all('a',class_="LatestNews-headline"):
    head_lines.append(i.text)
head_lines
```

# ii) Time

```python
time=[]
for j in soup.find_all('time',class_="LatestNews-timestamp"):
    time.append(j.text)
time
```

```python
links=[]
for k in soup.find_all('a',class_="LatestNews-headline"):
    links.append(k['href'])
links
```

```python
data= pd.DataFrame({'Time': time,'Headlines':head_lines,'Links':links})
data
```

# 6) Write a python program to scrape the details of most downloaded articles from AI in last 90 days.https://www.journals.elsevier.com/artificial-intelligence/most-downloaded-articles. Scrape below mentioned details and make data frame

# i) Paper Title ii) Authors iii) Published Date iv) Paper URL

```python
page = requests.get('https://www.journals.elsevier.com/artificial-intelligence/most-downloaded-articles')
soup = BeautifulSoup(page.content)
title=[]
url=[]
author=[]
publish=[]
for i in soup.find_all('a',class_="sc-5smygv-0 fIXTHm"):        
        url.append(i['href'])
        title.append(i.text)

for j in soup.find_all('span',class_="sc-1w3fpd7-0 dnCnAO"):
    author.append(j.text)
    
for k in soup.find_all('span',class_="sc-1thf9ly-2 dvggWt"):
    publish.append(k.text)

data=pd.DataFrame({"Paper Title": title,"Author Name": author, "Publish Date": publish, "Paper URL": url})
data
```

# 7) Write a python program to scrape mentioned details from dineout.co.inand make data frame

# i) Restaurant name
# ii) Cuisine
# iii) Location
# iv) Ratings
# v) Image URL

```python
page = requests.get('https://www.dineout.co.in/delhi-restaurants/buffet-special')
soup = BeautifulSoup(page.content)
```

```python
restaurent_name =[]
for rest in soup.find_all('a',class_="restnt-name ellipsis"):
    restaurent_name.append(rest.text)
```

```python
location = []
for loc in soup.find_all('div',class_="restnt-loc ellipsis"):
    location.append(loc.text)
```

```python
cuisine= []
for cus in soup.find_all('span',class_="double-line-ellipsis"):
    cuisine.append(cus.text.split('|')[1])
```

```python
ratings = []
for rat in soup.find_all('div',class_="restnt-rating rating-4"):
    ratings.append(rat.text)
```

```python
image_url=[]
for i in soup.find_all('img',class_="no-img"):
    image_url.append(i['data-src'])
```

```python
data= pd.DataFrame({'Restaurent Name': restaurent_name, 'Cusine':cuisine, 'Location': location, 'Ratings': ratings,'Images': image_url})
data
```
