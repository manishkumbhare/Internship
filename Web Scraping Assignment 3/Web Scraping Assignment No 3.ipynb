# 1. Write a python program which searches all the product under a particular product from www.amazon.in. The product to be searched will be taken as input from user. For e.g. If user input is ‘guitar’. Then search for guitars

```python
#importing all the required libraries
import selenium
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium.common.exceptions import NoSuchElementException
from selenium.webdriver.common.keys import Keys
import requests
import pandas as pd
import warnings
warnings.filterwarnings('ignore')
from time import sleep
import time
```

```python
#open Chrome browser
driver=webdriver.Chrome()
```

```python
#open webpage www.amazon.in
driver.get('https://www.amazon.in')
```

```python
#asking for input keyword to be searched
keyword=str(input('type the key word to be sarched: '))
```

```python
#search Keyword
search=driver.find_element(By.XPATH,'/html/body/div[1]/header/div/div[1]/div[2]/div/form/div[2]/div[1]/input')
search.send_keys(keyword)
```

```python
#click search button 
driver.find_element(By.XPATH,'/html/body/div[1]/header/div/div[1]/div[2]/div/form/div[3]/div/span/input').click()
time.sleep(5)
```

```python
#scraping the data for first 3 pages
prod_name=[]
URL=[]
last=driver.find_element(By.XPATH,'//span[@class="s-pagination-item s-pagination-disabled"]')
end=int(last.text)

for j in range(0,3):
    res=driver.find_elements(By.XPATH,'//a[@class="a-link-normal s-underline-text s-underline-link-text s-link-style a-text-normal"]')

    for i in res:
        try:
            prod_name.append(i.text)
        except:
            prod_name.append('-')

    RL=driver.find_elements(By.XPATH,'//a[@class="a-link-normal s-underline-text s-underline-link-text s-link-style a-text-normal"]')
    for i in RL:
        try:
            URL.append(i.get_attribute('href'))
            
        except:
            URL.append('-')
    ab = '/html/body/div[1]/div[2]/div[1]/div[1]/div/span[1]/div[1]/div[26]/div/div/span/a['
    cd = str(j+3)
    bc = str(ab + cd + ']')
    
    driver.find_element(By.XPATH,bc).click()
    time.sleep(10)

#calculating length of each list
print(len(prod_name),len(URL))
```

```python
#creating dataframe
df=pd.DataFrame({'Product Name':prod_name,'URL':URL})
df
```

```python
#scraping the data- comments , upvote and comment time
comments=[]
comment_time=[]
upvote=[]

try:
    comm='0'
    while (len(comm))<500:
        comm=driver.find_elements(By.XPATH,'//div[@class="style-scope ytd-expander"]/yt-formatted-string')
        for _ in range(0,20):
            driver.execute_script('window.scrollBy(0,100)')
        time.sleep(1)

except NoSuchElementException as e:
    print(e)
    
try:
    comm=driver.find_elements(By.XPATH,'//div[@class="style-scope ytd-expander"]/yt-formatted-string/span')
    for i in comm:
        comments.append(i.text)
except NoSuchElementException:
    comments.append('-')

try:
    ti=driver.find_elements(By.XPATH,'//div[@class="style-scope ytd-comment-renderer"]/div/div[2]/yt-formatted-string/a')
    for i in ti:
        comment_time.append(i.text)
except NoSuchElementException:
    comment_time.append('-')

try:
    upv=driver.find_elements(By.XPATH,'//span[@id="vote-count-middle"]')
    for i in upv:
        upvote.append(i.text)
except NoSuchElementException:
    upvote.append('-')

#getting the length of each list
print(len(comments),len(comment_time),len(upvote))

#creating the dataframe for first 500 billionairs
df=pd.DataFrame({'Comments':comments[:500],'Time of comment':comment_time[:500],'Up Vote':upvote[:500]})
df
```

```python
#opening the Chrome browser
driver=webdriver.Chrome()
```

```python
#opening website 'www.hostelworld.com' and serarching for hostels in London
driver.get('https://www.hostelworld.com')
time.sleep(5)
search=driver.find_element(By.XPATH,'/html/body/div[3]/div/div/div[2]/div[2]/div[2]/div/div/div/div[1]/div[1]/div/div[2]/input')
search.send_keys('London')
time.sleep(2)
driver.find_element(By.XPATH,'/html/body/div[3]/div/div/div[2]/div[2]/div[2]/div/div/div/div[1]/div[2]/div/ul/li[2]/button').click()
time.sleep(1)
driver.find_element(By.XPATH,'/html/body/div[3]/div/div/div[2]/div[2]/div[2]/div/div/div/div[5]/button[1]').click()
```

```python
#scraping the required data
urls=[]
hostel_name=[]
dist_from_city_centre=['-']
ratings=[]
total_reviews=[]
overall_reviews=[]
ficility=[]
prop_disc=[]
private_price=[]
droms_price=[]


try:
    url=driver.find_elements(By.XPATH,'//div[@class="property-card"]/a') #32
    for i in url:
        urls.append(i.get_attribute('href'))
except NoSuchElementException:
    urls.append('-')

try:
    dist=driver.find_elements(By.XPATH,'//span[@class="distance-description"]') #30
    for i in dist:
            dist_from_city_centre.append(i.text)   
except NoSuchElementException:
    dist_from_city_centre.append('-')

try:
    name=driver.find_elements(By.XPATH,'//div[@class="property-name"]') #32
    for i in name:
            hostel_name.append(i.text)   
except NoSuchElementException:
    hostel_name.append('-')

try:
    rating=driver.find_elements(By.XPATH,'//span[@class="number"]') #32
    for i in rating:
            ratings.append(i.text)   
except NoSuchElementException:
    ratings.append('-')

    
#opening each page for getting more data
for url in urls:
    driver.get(url)  
    time.sleep(1)
    try:
        over_rev=driver.find_element(By.XPATH,'/html/body/div[3]/div/div/div[2]/section/div[9]/div/div[1]/div[1]/div[2]/div[1]/span')
        overall_reviews.append(over_rev.text)
    except NoSuchElementException:
        overall_reviews.append('-')
        
    try:
        prop=driver.find_element(By.XPATH,'/html/body/div[3]/div/div/div[2]/section/div[9]/div/div[2]/div/div/div[2]')
        prop_disc.append(prop.text)        
    except NoSuchElementException:         
        prop_disc.append('-')              
    
    try:
        private=driver.find_element(By.XPATH,'/html/body/div[3]/div/div/div[2]/section/div[11]/div/div/section/div/div[1]/div/div[1]/div[2]/div[1]/div[2]/div[2]/div/div[1]/div[2]/div[1]')
        private_price.append(private.text)
    except NoSuchElementException:
        private_price.append('-')
    
    try:
        droms=driver.find_element(By.XPATH,'/html/body/div[3]/div/div/div[2]/section/div[11]/div/div/section/div/div/div[2]/div[2]/div[2]/div/div[2]/div[1]/div/div[1]/div[2]/div[1]')
        droms_price.append(droms.text)      
    except NoSuchElementException:
        droms_price.append('-')
    
    try:
        rev=driver.find_element(By.XPATH,'/html/body/div[3]/div/div/div[2]/section/div[9]/div/div[1]/div[1]/div[2]/div[2]')
        total_reviews.append(rev.text)      
    except NoSuchElementException:
        total_reviews.append('-')
    
    try:
        fic=driver.find_element(By.XPATH,'/html/body/div[3]/div/div/div[2]/section/div[13]/div/ul')
        ficility.append(fic.text)      
    except NoSuchElementException:
        ficility.append('-')

#print(len(urls),len(hostel_name),len(dist_from_city_centre),len(ratings),len(total_reviews),len(overall_reviews),len(ficility),len(prop_disc),len(private_price),len(droms_price))
```
