jobs data.

2. Enter ‚ÄúData Analyst‚Äù in ‚ÄúJob title, Skills‚Äù field and enter ‚ÄúBangalore‚Äù in ‚Äúenter the location‚Äù field.
3. Then click the searchbutton.

```python
#importing all required libraries
import selenium
from selenium import webdriver
from selenium.webdriver.common.by import By
import pandas as pd
import warnings
warnings.filterwarnings('ignore')
from time import sleep
import time
```

```python
#initializing chrome
driver=webdriver.Chrome()
```

```python
#open the desired URL
driver.get('https://www.shine.com')
```

```python
#Search & input job title
search=driver.find_element(By.XPATH,'/html/body/div[1]/header[1]/div[3]/div/div/div[1]/div/span/i')
search.click()
```

```python
#Search & input job title
profile=driver.find_element(By.XPATH,'/html/body/div[1]/div[4]/div/div[2]/div[2]/div/form/div/div[1]/ul/li[1]/div/input')
profile.send_keys('Data Analyst')
```

```python
#Search & Input job location
location=driver.find_element(By.XPATH,'/html/body/div[1]/div[4]/div/div[2]/div[2]/div/form/div/div[1]/ul/li[2]/div/input')
location.send_keys('Bangalore')
```

```python
#click Search button
search=driver.find_element(By.XPATH,'/html/body/div[1]/div[4]/div/div[2]/div[2]/div[1]/form/div/div[2]/div/button')
search.click()
```

```python
#Getting the first 10 jobs from search result
job_title=[]
job_location = []
company_name= []
experience_required = []
search_title=driver.find_elements(By.XPATH,'//h2[@itemprop="name"]')
for i in search_title[:10]:
    job_title.append(i.text)

search_location=driver.find_elements(By.XPATH,'//div[@class=" jobCard_jobCard_lists_item__YxRkV jobCard_locationIcon__zrWt2"]')
for i in search_location[:10]:
    job_location.append(i.text)

search_companyname=driver.find_elements(By.XPATH,'//div[@class="jobCard_jobCard_cName__mYnow"]')
for i in search_companyname[:10]:
    company_name.append(i.text)

search_experience=driver.find_elements(By.XPATH,'//div[@class=" jobCard_jobCard_lists_item__YxRkV jobCard_jobIcon__3FB1t"]')
for i in search_experience[:10]:
    experience_required.append(i.text)

#creating Dataframe
df=pd.DataFrame({"Job Title": job_title, "Job Location":job_location,"Company Name":company_name, "Experience Required": experience_required})
df
```

```python
#initializing chrome
driver=webdriver.Chrome()
```

```python
#opening the desired URL
driver.get('https://www.flipkart.com/apple-iphone-11-black-64-gb/product-reviews/itm4e5041ba101fd?pid=MOBFWQ6BXGJCEYNY&lid=LSTMOBFWQ6BXGJCEYNYZXSHRJ&market')
```

```python
#opening the desired URL
driver.get('https://www.azquotes.com/')
```

```python
#click on Top Quotes
driver.find_element(By.XPATH,'/html/body/div[1]/div[1]/div[1]/div/div[3]/ul/li[5]/a').click()
```

```python
#fetching the top 1000 quotes by scraping data of each page and clicking on next button
quote=[]
author=[]
type=[]

while (len(quote))<1000:
    
    qot=driver.find_elements(By.XPATH,'//a[@class="title"]')
    for i in qot:
        quote.append(i.text)

    aut=driver.find_elements(By.XPATH,'//div[@class="author"]')
    for i in aut:
        author.append(i.text)

    ty=driver.find_elements(By.XPATH,'//div[@class="tags"]')
    for i in ty:
        type.append(i.text)

    driver.find_element(By.XPATH,'/html/body/div[1]/div[2]/div/div/div/div[1]/div/div[3]/li[12]').click()
    
    #inserting delay of 2sec for proper opeing on page
    time.sleep(2)

#finding the length of each list
print(len(quote),len(author),len(type))

#creating Dataframe
df=pd.DataFrame({'Quotes':quote,'Author':author,'Type of Quote':type})
df
```
